```python
from ptt import PttCrawer
from flask import Flask, request, make_response, jsonify


app = Flask('flask-api')

@app.route('/')
def hello_word():
    message = {'message':'hello word'}
    return jsonify(message)

@app.route('/ptt',methods=['GET','POST'])
def run_ptt():
    
    if request.method =='GET':
        crawler=crawler=PttCrawer('https://www.ptt.cc/bbs/Gossiping/index.html',1)
    elif request.method=='POST':
        url=request.get_json().get('url','https://www.ptt.cc/bbs/Gossiping/index.html')
        page=request.get_json().get('page',1)
        
        crawler=crawler=PttCrawer(url,page)
        
    
    
    
    
    
    result=crawler.run()
    
    return jsonify(result)
```
